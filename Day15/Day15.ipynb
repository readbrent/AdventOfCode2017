```python
def generate_pair(prev1, prev2):
    new1 = (prev1 * 16807) % 2147483647
    new2 = (prev2 * 48271) % 2147483647
    return new1, new2

def isMatch(val1, val2):
    return bin(val1)[-16:] == bin(val2)[-16:]

def run_trials(seed1, seed2, num_trials):
    num_matches = 0
    first_input =  seed1
    second_input = seed2
    for _ in range(num_trials):
        first_input, second_input = generate_pair(first_input, second_input)
        
        if isMatch(first_input, second_input):
            num_matches += 1
            
    return num_matches


assert run_trials(1092455, 430625591, 5) == 1
```

```python
def generate1(seed):
    previous = seed
    while True:
        candidate = (previous * 16807) % 2147483647
        if candidate % 4 == 0:
            return candidate

        previous = candidate

def generate2(seed):
    previous = seed
    while True:
        candidate = (previous * 48271) % 2147483647
        if candidate % 8 == 0:
            return candidate
        else:
            previous = candidate
        
def run_trials2(seed1, seed2, num_pairs):
    gen1_state = seed1
    gen2_state = seed2
        
    for _ in range(num_pairs):
        gen1_state = generate1(gen1_state)
        gen2_state = generate2(gen2_state)
            
        if isMatch(gen1_state, gen2_state):
            num_matches += 1
                
    return num_matches 
```

```python
print(run_trials2(873, 583, 5000000))
```
