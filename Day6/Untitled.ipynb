```python
def cycle(memory_list):
    initial_index = memory_list.index(max(memory_list))
    index = initial_index + 1
    count = memory_list[initial_index]
    memory_list[initial_index] = 0
    while count:
        if index >= len(memory_list):
            index = 0
        count -= 1
        memory_list[index] += 1
        index += 1
    
    return memory_list
```

```python
def count_cycles(memory_list):    
    cycle_set = set()
    seen_once_set = set ()

    cycle_count = 0
    infinite_cycle_count = 0
    looped_once = False
    while True:
        str_mem = ''.join(str(memory_list))
        if str_mem in cycle_set:
            if str_mem in seen_once_set:
                print(infinite_cycle_count)
                return infinite_cycle_count
            else:
                seen_once_set.add(str_mem)
                print("Seen once!")
                looped_once = True
        cycle(memory_list)
        cycle_count += 1
        cycle_set.add(str_mem)
        
        if looped_once:
            print("Bumping the count!")
            infinite_cycle_count += 1
            print("Count is now " + str(infinite_cycle_count))
        
#count_cycles([0, 2, 7, 0])
count_cycles([14,0,15,12,11,11,3,5,1,6,8,4,9,1,8,4])
```
